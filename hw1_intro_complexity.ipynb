1. Comparison:

* $TA(n)=1000*n$
* $TB(n)=200*(n*log(n))$
* $TC(n)=10*n^{2}$
* $TD(n)=2^{n}$


```python
n=np.linspace(1,125)
```

```python
TA=1000*n
plt.figure(figsize=(3,3))
plt.plot(n,TA)
plt.title("TA(n)")
plt.xlabel("n")
plt.ylabel("T")
plt.show()
```

```python
TB=200*(n*np.log(n))
plt.figure(figsize=(3,3))
plt.plot(n,TB)
plt.title("TB(n)")
plt.xlabel("n")
plt.ylabel("T")
plt.show()
```

```python
TC=10*np.power(n,2)
plt.figure(figsize=(3,3))
plt.plot(n,TC)
plt.title("TC(n)")
plt.xlabel("n")
plt.ylabel("T")
plt.show()
```

```python
TD=np.power(2,n)
plt.figure(figsize=(3,3))
plt.plot(n,TD)
plt.title("TD(n)")
plt.xlabel("n")
plt.ylabel("T")
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.plot(n,TA,label="TA")
plt.plot(n,TB,label="TB")
plt.plot(n,TC,label="TC")
plt.plot(n,TD,label="TD")
plt.legend()
plt.xlabel("n")
plt.ylabel("T")
plt.show()
```

2. Rank the following complexities from highest to lowest

* $n*log(n)$
* $2^n$
* $n!$
* $n^{2}/2$
* $n^2 +log(n)$
* $n^2 *log(n)$
* $n + n^7 + n^3$
* $sqrt(n)$
* $(3/2)^n$
* $e^n$

```python
def nlogn(n):
    return n*np.log(n), "n*log(n)"

def two_to_n(n):
    return np.power(n,2), "2^n"

def fact(n):
    return scipy.special.factorial(n), "n!"

def n2_div2(n):
    return np.power(n,2) / 2, "(n^2)/2"

def n2_plus_logn(n):
    return np.power(n,2) + np.log(n), "n^2 + log(n)"

def n2logn(n):
    return np.power(n,2)*np.log(n), "(n^2)*log(n)"

def n_n7_n3(n):
    return n+np.power(n,7)+np.power(n,3), "n + n^7 + n^3"

def sqrtn(n):
    return np.sqrt(n), "sqrt(n)"

def three_halve_n(n):
    return np.power(3/2,n), "(3/2)^n"

def e_n(n):
    return np.exp(n), "e^n"
```

```python
functions=[nlogn(n),two_to_n(n),n2_div2(n),n2_plus_logn(n),n2logn(n),n_n7_n3(n),sqrtn(n),three_halve_n(n),e_n(n)]
```

```python
factorial=fact(n[:20])
```

```python
plt.figure(figsize=(8,8))
for func in functions:
    plt.plot(n, func[0],label=func[1])
plt.plot(n[:20],factorial[0],label=factorial[1])
plt.legend()
plt.xlabel("n")
plt.ylabel("T")
plt.ylim(10**1, 1000000000000000)
plt.yscale("log")
plt.grid(True)
plt.show()
```

By looking at the graph, the order (ascending) would be:
* $sqrt(n)$
* $n*log(n)$
* $n^{2}/2$
* $2^n$
* $n^2 +log(n)$
* $n^2 *log(n)$
* $n + n^7 + n^3$
* $(3/2)^n$
* $e^n$
* $n!$
